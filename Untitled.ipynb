# 1. Modeling the Leaky Integrate-and-Fire Neuron

The LIF neuron can be best understood as a passive RC circuit coupled to an active spike generator. The LIF neuron has two behavioral regimes: sub-threshold and super-threshold. In the sub-threshold regime, the LIF neuron behaves the same as a passive RC circuit. However, once the threshold is reached, the LIF neuron exhibits super-threshold behavior and fire a delta-function voltage spike. After firing, the neuron is then reset to the resting potential for some time period, ${\tau}_{ref}$, before it returns to the sub-threshold behavior.

![The Leaky Integrate-and-Fire Neuron](img/The Leaky Integrate-and-Fire Neuron.png)

## 1.1 Modeling the passive neuron
In the sub-threshold regime, the LIF neuron behaves the same as a passive RC circuit and can be described by the membrane equation: 
$$ \tau \frac{dV}{dt} = -(V - V_{rest}) + \frac{I_{inj}(t)}{G_{leak}}$$
where we define $\tau$ as: 
$$\tau = \frac{C}{G_leak}$$
To model the passive RC circuit, we adopt the following values for the membrane equation parametrs:
- Initial condition: $V(0) = V_{rest} = -70 mV$
- Membrane capacitance: $C = 100 pF$
- $G_{Leak} = 5 nS$.

```python
# initiate the membrane equation parameters
V_0 = -70            
C = 100                 
G_leak = 5   
tau = C/G_leak
```

We use Euler's rule to update the voltage according to the membrane equation (1). Then for a given step input current, we can plot the membrame voltage as a function of time:

```python
# define the membrane equation update rule
def membrane_equation_ODE(V, I_ti, V_0=V_0, G_leak=G_leak, tau=tau, dt=0.1):
    dV = (-(V - V_0) + I_ti/G_leak)/tau*dt
    V = V+dV
    return V
```

```python
# define an injected current
def I(t):               
  if t < 100: 
    return 0
  elif t < 500: 
    return 100
  else:
    return 0
```

```python
# find I(t) and V(t) for a given time interval 
t = np.linspace(0, 800, 8000) 
I_t = [I(i) for i in t]
V_t = np.zeros(len(t))
V_t[0] = V_0
for i in range(len(t)-1):
    V_t[i+1] = membrane_equation_ODE(V_t[i], I_t[i])
```

```python
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(8,6))
ax1.plot(t, I_t, 'b')
ax1.set_ylabel('injected current (pA)')
ax2.plot(t, V_t)
ax2.set_ylabel('membrane potential (mV)')
ax2.set_xlabel('time (ms)')
```

## 1.2 Modeling the LIF neuron
In the super-threshold regime,the LIF neuron is described by the additional threshold and reset mechanism, indicated by the components inside the dashed box in the figure below.
![LIF RC circuit](img/LIF RC circuit.png)
We define two additional parameters:
- Threshold: $V_{thresh} = -60 mV$
- Refractionary period: $\tau_{ref} = 5 ms $ 

```python
V_thresh = -55
tau_ref = 3 

def LIF_neuron(I_t, t, V_0=V_0, G_leak=G_leak, tau=tau, dt=0.1):
    V_t = np.zeros(len(t))
    V_t[0] = -70
    t_spikes = []
    for i in range(len(t)-1):   
        # if a spike just occurred within refractory period, then the neuron is stillstill resting 
        if  any(t_spike>t[i] - tau_ref for t_spike in t_spikes):
            V_t[i+1] = V_0
        # else, if the voltage is just arriving at the threshold, spike!
        elif V_t[i] >= V_thresh:
            V_t[i+1] = V_thresh + 10
            t_spikes.append(t[i+1])
        # else, just keep integrating the voltage same as a passive neuron
        else:
            V_t[i+1] = membrane_equation_ODE(V_t[i], I_t[i])
    return V_t, t_spikes
```

```python
V_t, t_spikes = LIF_neuron(I_t, t, V_0=V_0, G_leak=G_leak, tau=tau, dt=0.1)
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(8,6))
ax1.plot(t, I_t, 'b')
ax1.set_ylabel('injected current (pA)')
ax2.plot(t, V_t)
ax2.set_ylabel('membrane potential (mV)')
ax2.set_xlabel('time (ms)')
```

Great! Now we are seeing the expected behavior for LIF neurons. Notice that the membrane potential gradually returns to resting potential after the input current stops.
